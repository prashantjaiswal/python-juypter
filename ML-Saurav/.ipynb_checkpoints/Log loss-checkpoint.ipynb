```python
# -*- coding: utf-8 -*-

import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings("ignore")

p = np.arange(0,1,0.01)
p
```

```python
plt.plot(p,-1*np.log(p))
plt.show()

plt.plot(p, -1 * (np.log(1-p)))
plt.show()

plt.plot(p,-1*np.log(p))
plt.plot(p, -1 * (np.log(1-p)))
plt.show()
```
